```python
import pandas as pd
import numpy as np
```

```python
Employee=pd.DataFrame({'id':[1,2,3,4,5,6,7],'name':['Joe','Henry','Sam','Max','Janet','Randy','Will'],'salary':[85000,80000,60000,90000,69000,85000,70000],'departmentId':[1,2,2,1,1,1,1]})
Employee
```

```python
Department=pd.DataFrame({'id':[1,2],'name':['IT','Sales']})
Department
```

```python
#dense rank employee table by groupby
Employee['rank']=Employee.groupby('departmentId')['salary'].rank(method='dense',ascending=False)
Employee
```

```python
table=Employee.merge(Department,left_on='departmentId',right_on='id')
table
```

```python
result=table[table['rank']<=3][['name_y','name_x','salary']]
```

```python
result.rename['name_y':'Department','name_x':'Employee','salary':'Salary']
```
