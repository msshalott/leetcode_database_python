```python
import pandas as pd
import numpy as np
```

```python
Employee = pd.DataFrame({'id':[101,102,103,104,105,106],
                         'name':['John','Dan','James','Amy','Anne','Ron'],
                         'department':['A','A','A','A','A','B'],
                         'managerId':[np.nan,101,101,101,101,101]})
Employee
```

```python
employee_merge=Employee.merge(right=Employee,how='left',left_on='id',right_on='managerId')
employee_merge
```

```python
model = employee_merge.groupby('managerId_y')
model
```

```python
#This code filters the model DataFrame to keep only the groups where the length of the group (number of direct reports) is greater than or equal to 5. 
managers_with_five_or_more_reports = model.filter(lambda x: len(x) >= 5)['managerId_y'].unique()
result_table = Employee[Employee['id'].isin(managers_with_five_or_more_reports)][['name']].drop_duplicates()
result_table
```
